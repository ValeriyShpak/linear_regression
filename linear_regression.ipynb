## Построить линейную регрессионную модель. Прогнозирование цен на акции и получение среднеквадратичной ошибки и показателя R2

Ссылка на задание:

https://www.kaggle.com/abhimaneukj/tcs-dataset-from-2015-to-2021/tasks?taskId=6325

```python
import numpy as np

import pandas as pd

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error

from sklearn.metrics import r2_score

# Создадим датафрейм на основе csv-файла:

df_1 = pd.read_csv('TCS.NSE.csv', parse_dates=['Date'])
```

### Исходный датафрейм

```python
df_1
```

### Удалим строки с пропусками в столбцах 'Open', 'High', 'Low' и 'Close':

```python
new_df_1 = df_1.dropna(subset=['Open', 'High', 'Low', 'Close'])

new_df_1
```

### Подготовим регрессионную модель и вычислим RMSE и R2:

```python
# Объединим столбцы переменных в один столбец с помощью команды Numpy библиотеки np.c_. 
# А также создадим новую переменную для хранения целевых значений, указав фрейм данных и нужный столбец:

X = pd.DataFrame(np.c_[new_df_1['Open'], new_df_1['High'], new_df_1['Low']], columns=['Open', 'High', 'Low'])
                 
y = new_df_1['Close']

# Разделим фрейм данных на обучающие и тестовые наборы:

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

# Создадим экземпляр модели линейной регрессии и загоним в неё обучающие данные:

lin_reg_mod = LinearRegression()

lin_reg_mod.fit(X_train, y_train)

# Прогнозы на тестовом наборе будем делать с помощью команды "predict":

prediction = lin_reg_mod.predict(X_test)

# Проверим прогнозы относительно фактических значений с помощью метрик RMSE и R2:

test_set_rmse = (np.sqrt(mean_squared_error(y_test, prediction)))

test_set_r2 = r2_score(y_test, prediction)

print('rmse: ' + str(test_set_rmse))

print('r2: ' + str(test_set_r2))
```

### Значение r2 близко к 1, значит наша модель достаточно точна.
